## **Tujuan Metode EDA**
Metode **EDA (Exploratory Data Analysis)** digunakan untuk:
- Memahami struktur, distribusi, dan kualitas data.  
- Menemukan hubungan antar variabel (numerik dan kategorikal).  
- Mengidentifikasi nilai ekstrem (*outlier*) dan data yang hilang (*missing values*).  
- Melakukan pembersihan dan normalisasi data sebelum analisis lanjut.  
- Membandingkan karakteristik dan hasil analisis dari dua dataset yang serupa.  

---

## **Permasalahan yang Diselesaikan**
Dalam metode EDA ini, beberapa hal yang diselesaikan antara lain:
1. **Data Cleansing:**  
   Menghapus duplikasi, menormalkan nama kolom, dan memastikan format data seragam antara dua dataset.
2. **Data Integration:**  
   Menggabungkan dua dataset agar dapat dianalisis bersama untuk perbandingan.
3. **Analisis Statistik & Visualisasi:**  
   Menyajikan ringkasan deskriptif, distribusi nilai, dan hubungan antar variabel melalui grafik.
4. **Insight Akademik:**  
   Menemukan pola seperti pengaruh faktor sosial terhadap nilai siswa, distribusi nilai antar dataset, dan tingkat korelasi antar mata pelajaran.
5. **Data Enrichment:**  
   Membuat fitur baru (seperti rata-rata nilai) untuk memperdalam pemahaman pola performa siswa.  

---

## **Hasil yang Diharapkan**
- Dataset yang bersih dan siap dianalisis lebih lanjut.  
- Perbandingan menyeluruh antara dua dataset pendidikan.  
- Visualisasi yang menampilkan tren dan hubungan antar variabel.  
- Insight mendalam tentang faktor-faktor yang memengaruhi performa siswa.

## **5. Cek Nilai Anomali / Outlier pada Dataset 1 (df_student1)**
Pada tahap ini dilakukan pemeriksaan nilai anomali atau outlier pada Dataset 1 (Students Performance).
Outlier merupakan data yang nilainya menyimpang jauh dari pola umum â€” misalnya nilai ujian yang terlalu rendah atau terlalu tinggi dibanding siswa lain.
Deteksi ini penting agar analisis eksploratori data (EDA) tidak bias akibat adanya data ekstrem.

Langkah-langkah yang dilakukan:
1. Menampilkan statistik deskriptif untuk memahami rentang nilai tiap kolom numerik.
2. Menggunakan metode IQR (Interquartile Range) untuk mendeteksi data yang berada di luar batas wajar.
3. Menampilkan daftar kolom yang memiliki indikasi nilai outlier.

```python
# Tampilkan ringkasan statistik deskriptif
print("\nðŸ“ˆ Statistik Deskriptif Dataset 1:")
display(df_student1.describe())

# Pilih kolom numerik
numeric_cols_1 = df_student1.select_dtypes(include=['int64', 'float64']).columns

# Cek outlier menggunakan metode IQR
print("\nðŸ”Ž Deteksi Outlier (Metode IQR):")
for col in numeric_cols_1:
    Q1 = df_student1[col].quantile(0.25)
    Q3 = df_student1[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df_student1[(df_student1[col] < lower_bound) | (df_student1[col] > upper_bound)]
    if len(outliers) > 0:
        print(f"âš ï¸ Kolom '{col}' memiliki {len(outliers)} outlier.")
    else:
        print(f"âœ… Kolom '{col}' tidak memiliki outlier.")

print("\nðŸ” Pemeriksaan selesai untuk Dataset 1.\n")
```

Tahap ini memiliki tujuan yang sama seperti sebelumnya, yaitu untuk mendeteksi dan meninjau data ekstrem pada Dataset 2 (Students Performance 10K).
Karena dataset ini memiliki jumlah data lebih besar, potensi munculnya nilai aneh atau ekstrem juga lebih tinggi.

Langkah-langkah yang dilakukan:
1. Melihat ringkasan statistik semua kolom numerik untuk mendapatkan gambaran umum.
2. Menggunakan metode IQR (Interquartile Range) untuk mendeteksi nilai-nilai di luar batas normal.
3. Menampilkan kolom yang memiliki indikasi outlier untuk dilakukan peninjauan lebih lanjut pada tahap EDA.

```python
# Tampilkan ringkasan statistik deskriptif
print("\nðŸ“ˆ Statistik Deskriptif Dataset 2:")
display(df_student2.describe())

# Pilih kolom numerik
numeric_cols_2 = df_student2.select_dtypes(include=['int64', 'float64']).columns

# Cek outlier menggunakan metode IQR
print("\nðŸ”Ž Deteksi Outlier (Metode IQR):")
for col in numeric_cols_2:
    Q1 = df_student2[col].quantile(0.25)
    Q3 = df_student2[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df_student2[(df_student2[col] < lower_bound) | (df_student2[col] > upper_bound)]
    if len(outliers) > 0:
        print(f"âš ï¸ Kolom '{col}' memiliki {len(outliers)} outlier.")
    else:
        print(f"âœ… Kolom '{col}' tidak memiliki outlier.")

print("\nðŸ” Pemeriksaan selesai untuk Dataset 2.\n")
```

# **Menggabungkan Dataset**

```python
import pandas as pd

# --- Normalisasi Dataset 1 ---
student1_sel = df_student1.rename(columns={
    'sex': 'gender',
    'Medu': 'parental_level_of_education',
    'G1': 'math_score',
    'G2': 'reading_score',
    'G3': 'writing_score'
}).copy()
student1_sel['science_score'] = student1_sel[['math_score','reading_score','writing_score']].mean(axis=1)
student1_sel['total_score'] = student1_sel[['math_score','reading_score','writing_score']].sum(axis=1)
student1_sel['grade'] = student1_sel['total_score'].apply(lambda s: 'A' if s>=240 else 'B' if s>=180 else 'C' if s>=120 else 'D' if s>=60 else 'E')
student1_sel['dataset_origin'] = 'Students Performance Data Set'

# --- Normalisasi Dataset 2 ---
student2_sel = df_student2[['gender','parental_level_of_education','math_score','reading_score','writing_score',
                            'science_score','total_score','grade']].copy()
student2_sel['dataset_origin'] = 'Students Performance Dataset 10K'

# --- Gabungkan Kedua Dataset ---
common_cols = list(set(student1_sel.columns) & set(student2_sel.columns))
combined_students = pd.concat([student1_sel[common_cols], student2_sel[common_cols]], ignore_index=True)

# --- Hasil ---
print(f"âœ… Dataset berhasil digabung! Total baris: {len(combined_students)}, Total kolom: {len(common_cols)}")
display(combined_students.head())
```
